### 모듈 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
%matplotlib inline
```

### 데이터 불러오기, 변수 이름

```python
df = pd.read_csv('total.csv', index_col= '번호', encoding = 'utf-8') 
df.rename(columns = {'사진파일\n 이름':'이름', '함께 먹은 사람\n (연인,친구, 가족, 혼자)':'함께먹은사람',
                     '주문 금액\n (총 결제금액x)':'주문금액', '할인금액\n (쿠폰 등)':'할인금액'}, inplace = True)
df.head(3)
```

```python
pd.set_option('display.max_rows', 500)
df.iloc[150:200,10:]
```

### 오류 수정

```python
#439번 김현지 입력 오류 
df.loc[439,'주문날짜'] = '2012.12.31'
#627번 백수빈 입력 오류 
df.loc[627, '주문날짜'] = '2020.09.02'

#날짜 형태 변환 
df['주문날짜'] = pd.to_datetime(df['주문날짜'])
df['주문시간'] = pd.to_datetime(df['주문시간']) #시간 앞에 지금 날짜 들어감


#김현지 날짜 2012 로 오류
df = df[df.이름 != '김현지-1']

#주문금액에 쉼표 오류 : 23,000
df.loc[562,'주문금액'] = '23000'
df.loc[664,'주문금액'] ='15500'
df.loc[665,'주문금액'] ='15800'
df.loc[800,'주문금액'] = '20000'
#숫자형으로 
df['주문금액'] = pd.to_numeric(df['주문금액'])



#할인금액에 배민포인트 삭제 
df.drop(df[df['할인금액'] == '40(배민포인트)'].index, inplace = True)
#숫자형으로 
df['할인금액'] = pd.to_numeric(df['할인금액'])
```

# 시각화

1. 점포 주문횟수 / 주문금액 순위 
2. 점포 카테고리별 주문 평균/합계
3. 함께 먹은 사람 유형별 평균/합계 
4. 주문 날짜 --> 월별/계절별 주문금액 비교 
5. 결제 시간 --> 시간대별 주문금액 비교 

## 2. 점포 카테고리별 그래프

```python
df_category_sumt = df.groupby(by = '음식 카테고리').count()
df_category_sumt = df_category_sumt.sort_values('이름', ascending=False).reset_index()

plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothic'

sns.barplot(data = df_category_sumt,
            x = '음식 카테고리',
            y = '이름',
            capsize = 0.05)
plt.title('음식 카테고리별 주문 횟수')
plt.xticks(rotation = - 45 )
```

```python
df_category_sumt = df.groupby(by = '음식 카테고리').sum()
df_category_sumt = df_category_sumt.sort_values('주문금액', ascending=False).reset_index()

plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothic'


sns.barplot(data = df_category_sumt,
            x = '음식 카테고리',
            y = '주문금액',
            capsize = 0.05)
plt.title('음식 카테고리별 주문금액 합계')
plt.xticks(rotation = - 45 )
```

```python
df_category_sumt = df.groupby(by = '음식 카테고리').mean()
df_category_sumt = df_category_sumt.sort_values('주문금액', ascending=False).reset_index()

plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothic'

sns.barplot(data = df_category_sumt,
            x = '음식 카테고리',
            y = '주문금액',
            capsize = 0.05)
plt.title('음식 카테고리별 주문금액 평균')
plt.xticks(rotation = - 45 )
```

## 3. 함께 먹은 사람 별 그래프

```python
df_ppl_cnt = df.groupby(by = '함께먹은사람').count()
df_ppl_cnt = df_ppl_cnt.sort_values('이름', ascending=False).reset_index()

plt.figure(figsize=(10, 5))

sns.barplot(data = df_ppl_cnt,
            x = '함께먹은사람',
            y = '이름',
            capsize = 0.05)
plt.title('함께먹은 사람 별 주문횟수')
plt.xticks(rotation = - 45 )
```

```python
df_ppl_mean = df.groupby(by = '함께먹은사람').mean()
df_ppl_mean = df_ppl_mean.sort_values('주문금액', ascending=False).reset_index()

plt.figure(figsize=(10, 5))

sns.barplot(data = df_ppl_mean,
            x = '함께먹은사람',
            y = '주문금액',
            capsize = 0.05)
plt.title('함께먹은 사람 별 평균 주문금액')
plt.xticks(rotation = - 45 )
```
